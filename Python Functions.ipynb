```python
# In Python, both functions and methods are blocks of code designed to perform specific tasks, 
#but they differ in their association and usage within the language.
```

```python
#Functions:
```

```python
#Functions are independent blocks of code that can be called from anywhere in the program. 
#They are defined using the def keyword and are not tied to any object or class
```

```python
#Usage: Functions are called by their name and can be used to perform operations, return values, and promote code reusability.
#Example:

def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Output: Hello, Alice!
```

```python
#Methods are functions that are associated with an object or class. 
#They are defined within a class and require an instance of the class (or the class itself, in the case of class methods) to be called
```

```python
# Question 2)  Explain the concept of function arguments and parameters in Python.
```

```python
#In Python, functions are a fundamental part of the language, allowing you to encapsulate code for reuse and organization. 
#The concepts of arguments and parameters is crucial for effectively using functions.
```

```python
"""
Parameters vs. Arguments:
Parameters are the variables listed inside the parentheses in the function definition. 
They act as placeholders for the values that will be passed to the function when it is called.

Arguments are the actual values that are passed to the function when it is called. 
These values are assigned to the corresponding parameters.
"""
```

```python
#Example
#Here’s a simple example to illustrate the difference:
```

```python
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

greet("Alice")  # "Alice" is an argument
```

```python
#Types of Arguments
#1) Positional Arguments: These are the most common type of arguments. 
#They are assigned to parameters based on their position in the function call.

def add(a, b):
    return a + b

result = add(3, 5)  # 3 and 5 are positional arguments
print(result)  # Output: 8
```

```python
#2)Keyword Arguments: You can specify arguments by name, which allows you to pass them in any order.

def describe_pet(animal_type, pet_name):
    print(f"I have a {animal_type} named {pet_name}.")

describe_pet(pet_name="Whiskers", animal_type="cat")  # Using keyword arguments
```

```python
#3)Default Arguments: You can provide default values for parameters. 
#If an argument is not provided for that parameter, the default value is used.

def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()          # Output: Hello, Guest!
greet("Bob")    # Output: Hello, Bob!
```

```python
#Variable-length Arguments: You can use *args and **kwargs to allow a function to accept an arbitrary number of positional or keyword arguments.
#*args allows you to pass a variable number of positional arguments.

def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))  # Output: 10
```

```python
#**kwargs allows you to pass a variable number of keyword arguments.

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30, city="New York")
```

```python
"""
Iterable:
An iterable is any Python object that can return an iterator. 
It is an object that can be looped over (iterated over) using a for loop or any other iteration context. 
Common examples of iterables include lists, tuples, strings, dictionaries, and sets.

An iterable implements the __iter__() method, which returns an iterator.
"""
```

```python
"""
Iterator:
An iterator is an object that represents a stream of data. 
It is an object that implements two methods: __iter__() and __next__(). 
The __iter__() method returns the iterator object itself, 
while the __next__() method returns the next value from the iterator. 
When there are no more items to return, __next__() raises a StopIteration exception to signal that the iteration is complete.


"""
```

```python
"""
Key Differences
Definition:
Iterable: An object that can return an iterator.
Iterator: An object that represents a stream of data and implements the iterator protocol (__iter__() and __next__()).

Usage:
You can loop over an iterable using a for loop.
An iterator is used to fetch the values one at a time.

State:
An iterable can be looped over multiple times (it creates a new iterator each time).
An iterator maintains its state; once it is exhausted, it cannot be reused.
"""
```

```python
#Example 1: Iterable

# A simple list is an iterable
my_list = [1, 2, 3, 4, 5]

# You can iterate over it using a for loop
for item in my_list:
    print(item)
```

```python
#Example 2: Iterator

# Getting an iterator from an iterable
my_list = [1, 2, 3, 4, 5]
my_iterator = iter(my_list)  # Create an iterator from the iterable

# Using the iterator to get values
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
```

```python
#Question 6) Explain the concept of generators in Python and how they are defined.
```

```python
"""
Generators in Python are a special type of iterator that allow you to create iterators in a very efficient and concise way. 
They enable you to iterate through a sequence of values without having to store the entire sequence in memory at once. 
This is particularly useful when dealing with large datasets or streams of data.
"""
```

```python
"""
Key Features of Generators:
Memory Efficiency: Generators yield items one at a time and only when required, which means they use less memory compared to lists or other data structures that store all items at once.

Lazy Evaluation: Generators compute their values on-the-fly, which can lead to performance improvements, especially for large datasets.

Simpler Syntax: Generators can be defined using a simple function with the yield statement instead of return.
"""
```

```python
"""
Defining Generators
A generator can be defined using a function that contains one or more yield statements. 
When the function is called, it returns a generator object without starting execution immediately. 
Each time next() is called on the generator, the function executes until it hits a yield statement, which produces a value and suspends the function’s state. 
The next time next() is called, it resumes execution just after the yield.
"""
```

```python
#Example 1: Basic Generator
#Here’s a simple example of a generator that yields numbers from 0 to a specified maximum:
def count_up_to(max):
    count = 0
    while count < max:
        yield count
        count += 1

# Creating a generator object
counter = count_up_to(5)

# Using the generator
for number in counter:
    print(number)
```

```python
#Example 2: Generator Expression
#You can also create generators using generator expressions, which are similar to list comprehensions but use parentheses instead of square brackets:
```

```python
# Generator expression
squares = (x * x for x in range(5))

# Using the generator
for square in squares:
    print(square)
```

```python
#Example 3: Infinite Generator
#Generators can also be used to create infinite sequences. 
#Here’s an example of a generator that produces an infinite series of Fibonacci numbers:
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Using the infinite Fibonacci generator
fib = fibonacci()
for _ in range(10):
    print(next(fib))
```

```python
#Question 7) What are the advantages of using generators over regular functions?
```

```python
"""
1. Memory Efficiency
Generators do not store all values in memory at once; instead, they generate values on-the-fly. 
This is especially useful when dealing with large datasets or infinite sequences.
"""
#Example
def large_range(n):
    for i in range(n):
        yield i
# Using a generator
gen = large_range(1000000)

# This won't consume memory for all 1,000,000 integers at once
print(next(gen))
print(next(gen))
```

```python
"""
2. Lazy Evaluation
Generators evaluate their values only when required. 
This means you can start processing data without waiting for the entire dataset to be generated.
"""
def lazy_range(n):
    for i in range(n):
        yield i

# Using the generator
for number in lazy_range(10):
    print(number)  # Values are generated one at a time
```

```python
#In contrast, a regular function that returns a list would compute all values upfront:

def eager_range(n):
    return list(range(n))

# All values are computed and stored in memory at once
for number in eager_range(10):
    print(number)
```

```python
"""
3. Maintaining State
Generators maintain their state between calls, which allows them to remember where they left off. 
This is particularly useful for iterating through complex data structures or implementing custom iteration logic.
"""
#Example
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Using the generator
for count in countdown(5):
    print(count)
```

```python
#A regular function would not be able to maintain state in the same way without using global variables or returning complex data structures.
```

```python
#4. Infinite Sequences
#Generators can easily represent infinite sequences without running out of memory, as they generate values on demand.

def infinite_fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Using the generator
#A regular function would not be able to handle this without a predefined limit, as it would try to compute all values at once
fib = infinite_fibonacci()
for _ in range(10):
    print(next(fib))  # Will print the first 10 Fibonacci numbers
```

```python
#Question 8)  What is a lambda function in Python and when is it typically used?
```

```python
"""
A lambda function in Python is a small, anonymous function defined using the lambda keyword. 
Unlike regular functions defined using the def keyword, lambda functions can have any number of arguments but only a single expression. 
The expression is evaluated and returned when the lambda function is called.
"""
```

```python
#Syntax of a Lambda Function
#The syntax for a lambda function is as follows:
#lambda arguments: expression
```

```python
"""
* Characteristics of Lambda Functions
1)Anonymous: Lambda functions do not have a name unless assigned to a variable.
2)Single Expression: They can only contain a single expression, which is evaluated and returned.
3)Can Take Multiple Arguments: They can take multiple input arguments but still return a single value.
"""
```

```python
"""
* When to Use Lambda Functions
Lambda functions are typically used in situations where you need a simple function for a short period of time, 
often as an argument to higher-order functions (functions that take other functions as arguments). 

Common use cases include:

1)Sorting: When you want to sort a list of tuples or objects based on a specific criterion.
2)Filtering: When you want to filter elements from a list.
3)Mapping: When you want to apply a function to all items in a list.
4)Higher-order functions: In functions like map(), filter(), and reduce().
"""
```

```python
#1. Basic Example
#Here’s a simple example of a lambda function that adds two numbers:

add = lambda x, y: x + y
result = add(3, 5)
print(result)  # Output: 8
```

```python
#2. Using Lambda with map()
#You can use lambda functions with the map() function to apply a function to all items in a list:

numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16, 25]
```

```python
#3. Using Lambda with filter()
#Lambda functions can also be used with filter() to filter elements from a list:

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4, 6]
```

```python
#4. Using Lambda with reduce()
#You can also use lambda functions with reduce() from the functools module to apply a function cumulatively to the items of an iterable:

from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24
```

```python
#Question 9: Explain the purpose and usage of the `map()` function in Python.
```

```python
"""
The map() function in Python is a built-in higher-order function that allows you to apply a specified function to every item in an iterable (like a list, tuple, or string) and return a map object (which is an iterator). 
This can be particularly useful for transforming data in a concise and readable way.

Purpose of map()
Transformation: It is primarily used to transform data from one form to another by applying a function to each item in an iterable.
Conciseness: It allows you to write more concise and readable code compared to using loops.
Functional Programming: It fits well into a functional programming style, where functions are first-class citizens.
"""
```

```python
#Return Value
#The map() function returns a map object (an iterator). 
#You can convert it to a list or another iterable type if needed.
```

```python
#1. Basic Example
#Here's a simple example of using map() to square each number in a list:

numbers = [1, 2, 3, 4, 5]

# Define a function to square a number
def square(x):
    return x ** 2

# Use map to apply the square function to each element in the numbers list
squared_numbers = map(square, numbers)

# Convert the map object to a list
squared_list = list(squared_numbers)
print(squared_list)  # Output: [1, 4, 9, 16, 25]
```

```python
#2. Using Lambda Functions with map()
#You can also use lambda functions to make the code more concise:

numbers = [1, 2, 3, 4, 5]

# Use map with a lambda function to square each number
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert the map object to a list
squared_list = list(squared_numbers)
print(squared_list)  # Output: [1, 4, 9, 16, 25]
```

```python
#3. Mapping Multiple Iterables
#The map() function can take multiple iterables. 
#In this case, the function must take as many arguments as there are iterables.

list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Use map to add corresponding elements from two lists
result = map(lambda x, y: x + y, list1, list2)

# Convert the map object to a list
result_list = list(result)
print(result_list)  # Output: [5, 7, 9]
```

```python
#4. Converting Strings to Integers
#You can use map() to convert a list of strings to integers:

string_numbers = ['1', '2', '3', '4', '5']

# Use map to convert strings to integers
int_numbers = map(int, string_numbers)

# Convert the map object to a list
int_list = list(int_numbers)
print(int_list)  # Output: [1, 2, 3, 4, 5]
```

```python
#5. Using map() with Custom Functions
#You can define custom functions and use them with map() to perform more complex transformations.

def to_uppercase(s):
    return s.upper()

words = ['hello', 'world', 'python']

# Use map to convert each word to uppercase
uppercase_words = map(to_uppercase, words)

# Convert the map object to a list
uppercase_list = list(uppercase_words)
print(uppercase_list)  # Output: ['HELLO', 'WORLD', 'PYTHON']
```

```python
#1. map()
#Purpose: The map() function applies a specified function to each item in an iterable (like a list or tuple) and returns a new iterable (map object) containing the results.

#Syntax:
#map(function, iterable, ...)
```

```python
#Example
numbers = [1, 2, 3, 4, 5]

# Using map to square each number
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert to list
squared_list = list(squared_numbers)
print(squared_list)  # Output: [1, 4, 9, 16, 25]
```

```python
#2. filter()
#Purpose: The filter() function constructs an iterator from elements of an iterable for which a function returns true. 
#In other words, it is used to filter out elements based on a condition.

#Syntax:
#filter(function, iterable)
```

```python
#Example
numbers = [1, 2, 3, 4, 5, 6]

# Using filter to get even numbers
even_numbers = filter(lambda x: x % 2 == 0, numbers)

# Convert to list
even_list = list(even_numbers)
print(even_list)  # Output: [2, 4, 6]
```

```python
#3. reduce()
#Purpose: The reduce() function (found in the functools module) applies a rolling computation to sequential pairs of values in an iterable, effectively reducing the iterable to a single cumulative value.
#Syntax
#from functools import reduce
#reduce(function, iterable[, initializer])
```

```python
from functools import reduce

numbers = [1, 2, 3, 4]

# Using reduce to compute the product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24 (1 * 2 * 3 * 4)
```

```python
# Question 11: Write a Python program, which returns a list with 2-tuples. 
#Each tuple consists of the order number and the product of the price per item and the quantity. 
#The product should be increased by 10,- € if the value of the order is smaller than 100,00 €. 
#Write a Python program using lambda and map.
```

```python
# Sample data
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Function to calculate the adjusted total price
def calculate_total(order):
    order_number = order[0]
    quantity = order[2]
    price_per_item = order[3]
    
    total_price = quantity * price_per_item
    # Increase total by 10 if it's less than 100
    if total_price < 100:
        total_price += 10
    
    return (order_number, total_price)

# Using map to apply the calculate_total function to each order
result = list(map(calculate_total, orders))

# Output the result
print(result)
```
